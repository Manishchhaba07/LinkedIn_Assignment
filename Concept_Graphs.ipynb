```python
def EL_to_AL(edge):  
    AL = {}  # Initialize an empty dictionary to store adjacency list
    for node1, node2 in edge:  # Loop through each edge (pair of nodes)
        if node1 not in AL:  # If node1 is not already in the adjacency list
            AL[node1] = []  # Add node1 with an empty list
        AL[node1].append(node2)  # Add node2 to the adjacency list of node1

        if node2 not in AL:  # If node2 is not already in the adjacency list
            AL[node2] = []  # Add node2 with an empty list
        AL[node2].append(node1)  # Add node1 to the adjacency list of node2

    return AL  # Return the adjacency list

def AM_to_AL(Matrix):
    dict = {}  # Initialize an empty dictionary to store adjacency list
    for i in range(0, len(Matrix)):  # Loop through each row in the matrix
        dict[f"S{i+1}"] = []  # Create an empty list for each node
        for j in range(0, len(Matrix)):  # Loop through each column
            if Matrix[i][j] == 1:  # If there is an edge between i and j
                dict[f"S{i+1}"].append(f"S{j+1}")  # Add the corresponding node to the list
                
    return dict  # Return the adjacency list

def adjacent(graph, a, b):
    if type(graph) == list and type(graph[0]) == tuple:  # If the graph is in edge list format
        update_graph = EL_to_AL(graph)  # Convert edge list to adjacency list

        if a in update_graph[b]:  # Check if there's an edge from node b to node a
            return True
        return False
    
    elif type(graph) == list and type(graph[0]) == list:  # If the graph is in adjacency matrix format
        update_graph = AM_to_AL(graph)  # Convert adjacency matrix to adjacency list

        if a in update_graph[b]:  # Check if there's an edge from node b to node a
            return True
        return False
    
    else:  # If the graph is already in adjacency list format
        if a in graph[b]:  # Check if there's an edge from node b to node a
            return True
        return False

MyGraph1 = [('S1', 'S2'), ('S1', "S3"), ('S1', 'S5'), ('S2', 'S4'), ('S3', 'S5')]  # Edge list format

MyGraph2 = [
        [0, 1 , 1, 0, 1],  # Adjacency matrix format
        [1, 0 , 0, 1, 0],
        [1, 0 , 0, 0, 1],
        [0, 1 , 0, 0, 0],
        [1, 0 , 1, 0, 0]
    ]

MyGraph3 = {"S1":["S2",'S3', 'S5'],  # Adjacency list format
                  "S2":["S1",'S4'],
                  "S3":["S1", 'S5'],
                  "S4":["S2"],
                  'S5':['S1', 'S3']
                  } 

print(adjacent(MyGraph1,"S3","S1"))  # Test the function with edge list
```

```python
def EL_to_AL(edge):  
    AL = {}  # Initialize adjacency list as an empty dictionary
    for node1, node2 in edge:  # Iterate through each edge (pair of nodes)
        if node1 not in AL:
            AL[node1] = []  # Add node1 to the adjacency list if not present
        AL[node1].append(node2)  # Add node2 as a neighbor of node1

        if node2 not in AL:
            AL[node2] = []  # Add node2 to the adjacency list if not present
        AL[node2].append(node1)  # Add node1 as a neighbor of node2

    return AL  # Return the adjacency list

def AM_to_AL(Matrix):
    dict = {}  # Initialize adjacency list as an empty dictionary
    for i in range(0, len(Matrix)):  # Iterate through the rows (nodes)
        dict[f"S{i+1}"] = []  # Initialize an empty list for each node
        for j in range(0, len(Matrix)):  # Iterate through the columns (nodes)
            if Matrix[i][j] == 1:  # If there's an edge between i and j
                dict[f"S{i+1}"].append(f"S{j+1}")  # Add node j as a neighbor of node i
                
    return dict  # Return the adjacency list

def is_graph_complete(graphs):
    # Check if the graph is in edge list format
    if type(graphs) == list and type(graphs[0]) == tuple:
        updated_graphs = EL_to_AL(graphs)  # Convert edge list to adjacency list

        n = len(updated_graphs)  # Get the number of nodes
        for nodes in updated_graphs:  # Check each node
            # If a node doesn't have n-1 neighbors or has a self-loop, return False
            if len(updated_graphs[nodes]) != n-1 or nodes in updated_graphs[nodes]:
                return False
        return True  # Return True if all nodes are connected correctly
    
    # Check if the graph is in adjacency matrix format
    elif type(graphs) == list and type(graphs[0]) == list:
        updated_graphs = AM_to_AL(graphs)  # Convert adjacency matrix to adjacency list
    
        n = len(updated_graphs)  # Get the number of nodes
        for nodes in updated_graphs:  # Check each node
            # If a node doesn't have n-1 neighbors or has a self-loop, return False
            if len(updated_graphs[nodes]) != n-1 or nodes in updated_graphs[nodes]:
                return False
        return True  # Return True if all nodes are connected correctly
    
    # Check if the graph is already in adjacency list format
    else:
        n = len(graphs)  # Get the number of nodes
        for nodes in graphs:  # Check each node
            # If a node doesn't have n-1 neighbors or has a self-loop, return False
            if len(graphs[nodes]) != n-1 or nodes in graphs[nodes]:
                return False
        return True  # Return True if all nodes are connected correctly
    

MyGraph1 = [('S1', 'S2'), ('S1', "S3"), ("S2", "S3")]  # Edge list format

MyGraph2 = [  # Adjacency matrix format
        [0, 1 , 1, 0, 1],
        [1, 0 , 0, 1, 0],
        [1, 0 , 0, 0, 1],
        [0, 1 , 0, 0, 0],
        [1, 0 , 1, 0, 0]
    ]

MyGraph3 = {  # Adjacency list format
    "S1": ["S2", 'S3', 'S5'],
    "S2": ["S1", 'S4'],
    "S3": ["S1", 'S5'],
    "S4": ["S2"],
    'S5': ['S1', 'S3']
} 
    
print(is_graph_complete(MyGraph3))  # Test with MyGraph3 to check if it's a complete graph
```

# Question 6.

Given a graph and a list of vertices, check if it forms a walk, or a trail or a path or none of these.

```python
def EL_to_AL(edge):  
    AL = {}  # Convert edge list to adjacency list
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []  # Add node1 to adjacency list if not present
        AL[node1].append(node2)  # Add node2 as a neighbor of node1

        if node2 not in AL:
            AL[node2] = []  # Add node2 to adjacency list if not present
        AL[node2].append(node1)  # Add node1 as a neighbor of node2

    return AL  # Return the adjacency list

def AM_to_AL(Matrix):
    dict = {}  # Convert adjacency matrix to adjacency list
    for i in range(0, len(Matrix)):
        dict[f"S{i+1}"] = []  # Initialize empty list for each node
        for j in range(0, len(Matrix)):
            if Matrix[i][j] == 1:  # If there's an edge between nodes i and j
                dict[f"S{i+1}"].append(f"S{j+1}")  # Add node j as a neighbor of node i
                
    return dict  # Return the adjacency list

def Graph_conversion(MyGraph):
    # Convert graph from edge list or adjacency matrix to adjacency list format
    if type(MyGraph) == list and type(MyGraph[0]) == tuple:
        return EL_to_AL(MyGraph)  # Convert edge list to adjacency list
    elif type(MyGraph) == list and type(MyGraph[0]) == list:
        return AM_to_AL(MyGraph)  # Convert adjacency matrix to adjacency list
    else:
        return MyGraph  # If already in adjacency list, return as is
        
def is_walk(graph, walk):
    # Check if the given sequence of nodes forms a valid walk in the graph
    for i in range(len(walk) - 1):
        if walk[i+1] not in graph[walk[i]]:  # Check if consecutive nodes are connected
            return False  # Return False if any pair of consecutive nodes is not connected
    return True  # Return True if all consecutive nodes are connected

def check_walk(walk):
    # Check if the walk is a Path, Trail, or not a valid path or trail
    repeated_nodes = []  # List to track repeated nodes
    repeated_edges = []  # List to track repeated edges
    for node in walk:
        if walk.count(node) > 1 and node not in repeated_nodes:  # Check for repeated nodes
            repeated_nodes.append(node)

    for i in range(len(walk) - 1):
        edge = (walk[i], walk[i + 1])  # Edge between consecutive nodes
        reverse_edge = (walk[i + 1], walk[i])  # Reverse edge

        count = 0
        for j in range(len(walk) - 1):  # Count the occurrences of edges
            current_edge = (walk[j], walk[j + 1])
            current_reverse_edge = (walk[j + 1], walk[j])

            if current_edge == edge or current_edge == reverse_edge:
                count += 1

        # If any edge occurs more than once, it's not a Trail or Path
        if count > 1 and edge not in repeated_edges and reverse_edge not in repeated_edges:
            repeated_edges.append(edge)

    # Determine the type of the walk
    if repeated_edges:
        return "Not a Trail, Not a Path"  # If there are repeated edges
    elif repeated_nodes:
        return "Trail"  # If there are repeated nodes but no repeated edges
    else:
        return "Path"  # If there are no repeated nodes or edges
    
def analyze_walk(graph, walk):
    # Analyze if the walk is valid and its type (Path, Trail, etc.)
    graph = Graph_conversion(graph)  # Convert the graph to adjacency list
    if is_walk(graph, walk):  # Check if the walk is valid
        print("It is a valid WALK.")
        type_walk = check_walk(walk)  # Check the type of the walk
        return type_walk
    else:
        return "Not a Valid Walk."  # Return if the walk is not valid
    

MyGraph1 = [('S1', 'S2'), ('S1', "S3"), ('S1', 'S5'), ('S2', 'S4'), ('S3', 'S5')]  # Edge list format

MyGraph2 = [  # Adjacency matrix format
        [0, 1 , 1, 0, 1],
        [1, 0 , 0, 1, 0],
        [1, 0 , 0, 0, 1],
        [0, 1 , 0, 0, 0],
        [1, 0 , 1, 0, 0]
    ]
    
MyGraph3 = {  # Adjacency list format
    "S1": ["S2", 'S3', 'S5'],
    "S2": ["S1", 'S4'],
    "S3": ["S1", 'S5'],
    "S4": ["S2"],
    'S5': ['S1', 'S3']
} 
    
given_walk = ["S1","S3","S5","S1"]  # Walk to analyze
print(analyze_walk(MyGraph2, given_walk))  # Analyze the walk for MyGraph2
```

Check if a given graph is a tree.

```python
def EL_to_AL(edge):  
    AL = {}  # Initialize an empty adjacency list
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []  # Add node1 to adjacency list if not present
        AL[node1].append(node2)  # Add node2 as a neighbor of node1

        if node2 not in AL:
            AL[node2] = []  # Add node2 to adjacency list if not present
        AL[node2].append(node1)  # Add node1 as a neighbor of node2
    return AL  # Return the adjacency list

def AM_to_AL(Matrix):
    dict = {}  # Initialize an empty dictionary for adjacency list
    for i in range(len(Matrix)):  # Iterate over rows of the matrix
        dict[f"S{i+1}"] = []  # Add a node with an empty list
        for j in range(len(Matrix)):  # Iterate over columns
            if Matrix[i][j] == 1:  # If there is an edge
                dict[f"S{i+1}"].append(f"S{j+1}")  # Add a neighbor to the node
    return dict  # Return the adjacency list

def Graph_conversion(MyGraph):
    # Convert the graph into an adjacency list format (from edge list or adjacency matrix)
    if type(MyGraph) == list and type(MyGraph[0]) == tuple:
        return EL_to_AL(MyGraph)  # Convert edge list to adjacency list
    elif type(MyGraph) == list and type(MyGraph[0]) == list:
        return AM_to_AL(MyGraph)  # Convert adjacency matrix to adjacency list
    else:
        return MyGraph  # Return if already in adjacency list format

def is_graph_connected(MyGraph):
    # Check if the graph is connected
    MyGraph = Graph_conversion(MyGraph)  # Convert the graph to adjacency list
    nodes = list(MyGraph.keys())  # Get the list of nodes in the graph
    visited_list = []  # List to keep track of visited nodes
    queue = [nodes[0]]  # Start BFS with the first node

    while queue:  # While there are nodes to visit
        node = queue.pop(0)  # Pop the first node in the queue
        if node not in visited_list:
            visited_list.append(node)  # Mark the node as visited
            for adj in MyGraph[node]:  # Check all adjacent nodes
                if adj not in visited_list and adj not in queue:
                    queue.append(adj)  # Add unvisited adjacent nodes to the queue

    # Check if all nodes have been visited (graph is connected if true)
    is_connected = len(visited_list) == len(nodes)
    
    return is_connected, MyGraph, visited_list  # Return whether the graph is connected

def has_cycle(graph, current, visited, parent):
    # Check if the graph contains a cycle
    visited.add(current)  # Mark the current node as visited
    for neighbor in graph[current]:  # Visit all neighbors
        if neighbor not in visited:
            if has_cycle(graph, neighbor, visited, current):  # Recursively check for cycles
                return True
        elif neighbor != parent:  # If a neighbor is visited but not the parent, there's a cycle
            return True
    return False  # Return False if no cycle is detected

def is_tree(MyGraph):
    # Check if the graph is a tree
    is_connected, graph, visited_nodes = is_graph_connected(MyGraph)  # Check if the graph is connected

    if not is_connected:  # If the graph is not connected, it can't be a tree
        return "Graph is not connected, so it's not a tree."

    visited = set()  # Set to keep track of visited nodes
    start_node = list(graph.keys())[0]  # Start from any node

    if has_cycle(graph, start_node, visited, None):  # Check if there is a cycle
        return "Graph has a cycle, so it's not a tree."

    edge_count = sum(len(neighbors) for neighbors in graph.values()) // 2  # Count edges
    node_count = len(graph)  # Count nodes

    # Check if the graph has exactly n-1 edges for n nodes (tree condition)
    if edge_count != node_count - 1:
        return "Graph doesn't have n-1 edges, so it's not a tree."

    return "Graph is a Tree"  # If all conditions are met, the graph is a tree

# Sample graphs for testing
MyGraph1 = [('S1', 'S2'), ('S1', "S3"), ("S2", "S3")]  
MyGraph2 = [
    [0, 1 , 1, 0, 1],
    [1, 0 , 0, 1, 0],
    [1, 0 , 0, 0, 1],
    [0, 1 , 0, 0, 0],
    [1, 0 , 1, 0, 0]
] 
MyGraph3 = {
    "S1": ["S2",'S3', 'S5'],
    "S2": ["S1",'S4'],
    "S3": ["S1", 'S5'],
    "S4": ["S2"],
    'S5': ['S1', 'S3']
}  
MyGraph4 = [('S1', 'S2'), ('S2', 'S3'), ('S3', 'S4')]  # This is a tree (no cycle, n-1 edges)

print(is_tree(MyGraph4))  # Check if MyGraph4 is a tree
```

Given a connected cyclic graph, find its spanning tree.

```python
def EL_to_AL(edge):  
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []
        AL[node1].append(node2)

        if node2 not in AL:
            AL[node2] = []
        AL[node2].append(node1)
    return AL

def AM_to_AL(Matrix):
    dict = {}
    for i in range(len(Matrix)):
        dict[f"S{i+1}"] = []
        for j in range(len(Matrix)):
            if Matrix[i][j] == 1:
                dict[f"S{i+1}"].append(f"S{j+1}")
    return dict

def Graph_conversion(MyGraph):
    if type(MyGraph) == list and type(MyGraph[0]) == tuple:
        return EL_to_AL(MyGraph)
    elif type(MyGraph) == list and type(MyGraph[0]) == list:
        return AM_to_AL(MyGraph)
    else:
        return MyGraph  # Already adjacency list

def find_spanning_tree(MyGraph):
    graph = Graph_conversion(MyGraph)
    visited = set()
    tree_edges = []

    def dfs(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                tree_edges.append((node, neighbor))  # Add edge to spanning tree
                dfs(neighbor)

    # Start from the first node
    start_node = list(graph.keys())[0]
    dfs(start_node)

    return tree_edges


# Test Graphs
MyGraph1 = [('S1', 'S2'), ('S1', 'S3'), ('S2', 'S3')]  # Edge List
MyGraph2 = [  # Adjacency Matrix
    [0, 1 , 1, 0, 1],
    [1, 0 , 0, 1, 0],
    [1, 0 , 0, 0, 1],
    [0, 1 , 0, 0, 0],
    [1, 0 , 1, 0, 0]
]
MyGraph3 = {  # Adjacency List
    "S1": ["S2", "S3", "S5"],
    "S2": ["S1", "S4"],
    "S3": ["S1", "S5"],
    "S4": ["S2"],
    "S5": ["S1", "S3"]
}


# Run Tests
print("Spanning Tree of MyGraph1 (Edge List):", find_spanning_tree(MyGraph1))
print("Spanning Tree of MyGraph2 (Adj Matrix):", find_spanning_tree(MyGraph2))
print("Spanning Tree of MyGraph3 (Adj List):", find_spanning_tree(MyGraph3))
```

# Question 11.
Given a tree and a node, find its height.

```python
# Convert Edge List to Adjacency List
def EL_to_AL(edge):  
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []
        AL[node1].append(node2)

        if node2 not in AL:
            AL[node2] = []
        AL[node2].append(node1)
    return AL

# Convert Adjacency Matrix to Adjacency List
def AM_to_AL(Matrix):
    dict = {}
    for i in range(len(Matrix)):
        dict[f"S{i+1}"] = []
        for j in range(len(Matrix)):
            if Matrix[i][j] == 1:
                dict[f"S{i+1}"].append(f"S{j+1}")
    return dict

# General graph conversion function
def Graph_conversion(MyGraph):
    if type(MyGraph) == list and type(MyGraph[0]) == tuple:
        return EL_to_AL(MyGraph)
    elif type(MyGraph) == list and type(MyGraph[0]) == list:
        return AM_to_AL(MyGraph)
    else:
        return MyGraph  # Already adjacency list

# Recursive function to find the height of a node
def find_node_height(graph, node, visited):
    # Mark the current node as visited
    visited.add(node)

    # Get the children (neighbors) of the node
    children = graph[node]

    # If there are no children, this is a leaf node
    if not children:
        return 0  # The height of a leaf node is 0

    # Calculate the height of each child and return the max height + 1
    max_height = 0
    for child in children:
        if child not in visited:
            child_height = find_node_height(graph, child, visited)
            max_height = max(max_height, child_height)

    # Return the height of the node (max of children's height + 1)
    return max_height + 1

# Function to find the height of a specific node
def get_height_of_node(MyGraph, node):
    graph = Graph_conversion(MyGraph)

    # We need to track visited nodes to avoid cycles
    visited = set()

    # Call the recursive function
    return find_node_height(graph, node, visited)

# Test Graphs
MyGraph1 = [('S1', 'S2'), ('S1', 'S3'), ('S2', 'S4')]  # Edge List
MyGraph2 = {  # Adjacency List
    "S1": ["S2", "S3"],
    "S2": ["S1", "S4"],
    "S3": ["S1"],
    "S4": ["S2"]
}

# Test: Find height of node 'S1' in MyGraph1 and MyGraph2
print("Height of node S2 in MyGraph1:", get_height_of_node(MyGraph1, 'S2'))
print("Height of node S1 in MyGraph2:", get_height_of_node(MyGraph2, 'S1'))
```

# Question 12.
Given a tree and a node, find its depth.

```python
# Convert Edge List to Adjacency List
def EL_to_AL(edge):  
    AL = {}
    for node1, node2 in edge:
        if node1 not in AL:
            AL[node1] = []
        AL[node1].append(node2)

        if node2 not in AL:
            AL[node2] = []
        AL[node2].append(node1)
    return AL

# Convert Adjacency Matrix to Adjacency List
def AM_to_AL(Matrix):
    dict = {}
    for i in range(len(Matrix)):
        dict[f"S{i+1}"] = []
        for j in range(len(Matrix)):
            if Matrix[i][j] == 1:
                dict[f"S{i+1}"].append(f"S{j+1}")
    return dict

# General graph conversion function
def Graph_conversion(MyGraph):
    if type(MyGraph) == list and type(MyGraph[0]) == tuple:
        return EL_to_AL(MyGraph)
    elif type(MyGraph) == list and type(MyGraph[0]) == list:
        return AM_to_AL(MyGraph)
    else:
        return MyGraph  # Already adjacency list

# Recursive function to find the depth of a node
def find_node_depth(graph, node, target, depth, visited):
    # Mark the current node as visited
    visited.add(node)

    # If the target node is found, return the current depth
    if node == target:
        return depth

    # Explore the children (neighbors) of the current node
    for neighbor in graph[node]:
        if neighbor not in visited:
            # Recur for the neighbor node
            result = find_node_depth(graph, neighbor, target, depth + 1, visited)
            if result != -1:
                return result  # If the node is found, return its depth

    return -1  # If the target node is not found in the tree

# Function to find the depth of a specific node
def get_depth_of_node(MyGraph, target):
    graph = Graph_conversion(MyGraph)

    # We need to track visited nodes to avoid cycles
    visited = set()

    # Call the recursive function starting from the root node (first node in the graph)
    root_node = list(graph.keys())[0]
    return find_node_depth(graph, root_node, target, 0, visited)

# Test Graphs
MyGraph1 = [('S1', 'S2'), ('S1', 'S3'), ('S2', 'S4')]  # Edge List
MyGraph2 = {  # Adjacency List
    "S1": ["S2", "S3"],
    "S2": ["S1", "S4"],
    "S3": ["S1"],
    "S4": ["S2"]
}

# Test: Find depth of node 'S4' in MyGraph1 and MyGraph2
print("Depth of node S4 in MyGraph1:", get_depth_of_node(MyGraph1, 'S4'))
print("Depth of node S4 in MyGraph2:", get_depth_of_node(MyGraph2, 'S4'))
```
